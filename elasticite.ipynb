```python
Ux_maxes = []
Uy_maxes = []
SVM_maxes = []

# Test for different poisson coefficients
v_tab = [.0, 0.3, 0.49]

for v in v_tab:
    ElasticitySolver_maille = ElasticitySolver(maille=2, v=v)
    ElasticitySolver_maille.solve()
    n_nodes = ElasticitySolver_maille.n_nodes
    results = ElasticitySolver_maille.results
    Ux_max = results['U'][:n_nodes].max()
    Uy_max = results['U'][n_nodes:].max()
    SVM_max = results['SVM'].max()

    Ux_maxes.append(Ux_max)
    Uy_maxes.append(Uy_max)
    SVM_maxes.append(SVM_max)

```
