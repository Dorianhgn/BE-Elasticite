```python
ElasticitySolver1 = ElasticitySolver()
```

```python
ElasticitySolver1.solve()
```

```python
ElasticitySolver1.plot_results()
```

```python
U_maxes = []
SVM_maxes = []

for maille in range(1,5):
    ElasticitySolver_maille = ElasticitySolver(maille=maille)
    ElasticitySolver_maille.solve()
    n_nodes = ElasticitySolver_maille.n_nodes
    results = ElasticitySolver_maille.results
    Ux_max = results['U'][:n_nodes].max()
    SVM_max = results['SVM'].max()

    U_maxes.append(Ux_max)
    SVM_maxes.append(SVM_max)

```
